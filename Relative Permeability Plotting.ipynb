```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sw = np.linspace(0,1,100)
```

```python
sw
```

```python
krw = sw**3
krw
```

```python
kro = (1-sw)**2
kro
```

```python
df = pd.DataFrame({'Water Saturation': sw, 'Rel Perm Water': krw, 'Rel Perm Oil': kro})
```

```python
df
```

```python
df.plot()
```

```python
df.plot(subplots = True)
```

```python
plt.plot(sw,krw)
```

```python
#step 2: kro vs sw
```

```python
plt.plot(sw,kro)
```

```python
# plotting kro and krw vs sw plot
```

```python
plt.plot(sw,krw)
plt.plot(sw,kro)
```

```python
# step 4: Giving names to the axis
```

```python
plt.plot(sw,kro)
plt.plot(sw,krw)

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilities")
```

```python
#step 5: Making a grid
```

```python
plt.plot(sw,kro)
plt.plot(sw,krw)

plt.xlabel("Water Saturation")
plt.ylabel("Relative permeabilities")

plt.grid(True)
```

```python
# Step 6: Adding legends to the plot
```

```python
plt.plot(sw,krw, label = 'Water Rel Perm')
plt.plot(sw,kro, label = 'Oil Rel Perm')

plt.xlabel('Water Saturation')
plt.ylabel('Relative Permeability')

plt.grid(True)
plt.legend()
```

```python
# step 7: Change the size of plot
```

```python
plt.figure(figsize = (12,8))
plt.plot(sw,krw, label = 'Water Rel Perm')
plt.plot(sw,kro, label = 'Oil Rel Perm')

plt.xlabel('Water Saturation')
plt.ylabel('Relative Permeability')

plt.grid(True)
plt.legend()
```

```python
#step 8: Change the line colour
```

```python

plt.plot(sw,krw, label = 'Water Rel Perm', color = 'blue')
plt.plot(sw,kro, label = 'Oil Rel Perm', color = 'green')

plt.xlabel('Water Saturation')
plt.ylabel('Relative Permeability')

plt.grid(True)
plt.legend()
```

```python
# step 10: Change the line thickness
```

```python
plt.plot(sw,krw, label = 'Water Rel Perm', color = 'blue',lw = 2)
plt.plot(sw,kro, label = 'Oil Rel Perm', color = 'green', lw = 6)

plt.xlabel('Water Saturation')
plt.ylabel('Relative Permeability')

plt.grid(True)
plt.legend()
```

```python
# Step 11: Adding title
```

```python
plt.plot(sw,krw, label = 'Water Rel Perm', color = 'blue')
plt.plot(sw,kro, label = 'Oil Rel Perm', color = 'green')

plt.xlabel('Water Saturation')
plt.ylabel('Relative Permeability')

plt.grid(True)
plt.legend()
plt.title('Relative Permeability Plots')
```

```python
#step 12: Changing the line structure
```

```python
plt.plot(sw,krw,'--', label = 'Water Rel Perm', color = 'blue')
plt.plot(sw,kro, label = 'Oil Rel Perm', color = 'green')

plt.xlabel('Water Saturation')
plt.ylabel('Relative Permeability')

plt.grid(True)

plt.legend()
plt.title('Relative Permeability Plots')
```
